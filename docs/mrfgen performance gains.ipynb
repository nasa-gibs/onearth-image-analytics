```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
astgtm = [[501.415, 496.02, 6.215],
[182.931, 498.349, 6.291],
[128.138, 512.749, 23.535],
[149.913, 523.639, 45.111],
[351.651, 552.984, 236.55599999999998]]

astgtm = np.array(astgtm).T
labels = ["real", "user", "sys"]
cores = [1, 3, 5, 7, 9]
```

```python
df = pd.DataFrame()
```

```python
for i, row in enumerate(astgtm):
    plt.plot(cores, row, label = labels[i])

plt.gca().set(title="Runtime for ASTGTM2 Product for different numbers of cores", xlabel="Num CPUs", ylabel="Time (s)")
plt.legend()
plt.show()
```
